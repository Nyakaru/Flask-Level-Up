The dataframe `mpg_df` loads the dataset that contains information on various car makes and models over a number years along with information on their engine specifications, including their MPG (miles-per-gallon) rating for both `cty` (city driving) and `hwy` (highway driving)

 4. Which is the most fuel efficient car in the `compact` class in the whole dataset?
 5. Which is the most fuel efficient 6 cylinder car make/model made after 2003?
 6. Which is Subaru's least fuel efficient model in the whole dataset?
 7. Which is the worst fuel efficient model/make in the `suv` class?


```python
# YOUR SECTION B ANSWER HERE (please use comments to number each of your answers)


###### Question 1.
# > number of rows of the dataframe
print('{:,}'.format(len(mpg_df.index))) # 234

###### Question 2.
# >names of all the columns of the dataframe
list(mpg_df.columns.values) 
            # ['manufacturer', 'model', 'displ', 'year', 'cyl', 'trans', 'drv', 'cty', 'hwy', 'fl', 'class']

###### Question 3.
# > 20th row of the dataframe
mpg_df.loc[19]
            # manufacturer             chevrolet
            # model           c1500 suburban 2wd
            # displ                          5.3
            # year                          2008
            # cyl                              8
            # trans                     auto(l4)
            # drv                              r
            # cty                             14
            # hwy                             20
            # fl                               r
            # class                          suv
            # Name: 19, dtype: object

###### Question 4.
# > most fuel efficient car in the compact class in the whole dataset
# create an avg_consumption ccolumn to store the average fuel consuption for city and highway driving
mpg_df['avg_consuption'] = mpg_df[['cty', 'hwy']].mean(axis=1)
mpg_df.loc[(mpg_df['class'] == 'compact') & \
(mpg_df['avg_consuption'] == mpg_df.loc[mpg_df['class'] == 'compact']['avg_consuption'].min())]
# Hence 1999, 4 cylinder  volkswagen jetta is the most efficient compact class car with 19.5 MPG consumption


###### Question 5.
# > the most fuel efficient 6 cylinder car make/model made after 2003
mpg_df.loc[(mpg_df['cyl'] == 6 ) &(mpg_df['year'] > 2003) &\
           (mpg_df['avg_consuption'] == mpg_df.loc[(mpg_df['cyl'] == 6 ) &(\
                                                   mpg_df['year'] > 2003)]['avg_consuption'].min())]['model']
# Hence caravan 2wd is the most fuel efficient
# 6 cylinder car make/model made after 2003 with 14.0 MPG consumption



###### Question 6.
# > Subaru's least fuel efficient model in the whole dataset
mpg_df.loc[(mpg_df['manufacturer'] == 'subaru') &\
           (mpg_df['avg_consuption'] == mpg_df.loc[(mpg_df['manufacturer'] == 'subaru')]\
            ['avg_consuption'].max())]['model']
# Hence forester awd and impreza awd are the most fuel inefficient
# Subaru models 23.5 MPG consumption


###### Question 7.
mpg_df.loc[(mpg_df['class'] == 'suv') & \
(mpg_df['avg_consuption'] == mpg_df.loc[mpg_df['class'] == 'suv']['avg_consuption'].max())]['model']
# Hence Subaru forester awd is the most fuel inefficient
# suv class with 23.5 MPG consumption
```
